```python
# notebooks/02_exploratory_analysis.ipynb

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger les donnes
data = pd.read_csv('../data/raw/synthetic_house_data.csv')

# Matrice de corrlation
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Matrice de corrlation des variables")
plt.show()

```

```python
# Distribution du prix de vente
plt.figure(figsize=(8, 6))
sns.histplot(data['SalePrice'], bins=50, kde=True)
plt.title('Distribution des prix de vente')
plt.show()

```
